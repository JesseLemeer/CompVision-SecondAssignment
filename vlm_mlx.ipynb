```python
import subprocess
```

```python
VIDEO_PATH = "TVSum/video_5.mp4"

PROMPT = (
    "Describe all the key events in the video." 
)

SYSTEM = (
    "Focus only on describing the key dramatic action or notable event occurring in this video segment. "
    "Skip general context or scene-setting details unless they are crucial to understanding the main action. "
    "Always format your response exactly as: "
    "Salient event 1: ... Salient event 2: ... etc. "
    "Do not use any other format."
)
```

```python
#Retrieval salient events
result = subprocess.run([
    "python", "-m", "mlx_vlm.smolvlm_video_generate",
    "--model", "mlx-community/SmolVLM2-256M-Video-Instruct-mlx",
    "--system", SYSTEM,
    "--prompt", PROMPT,
    "--video", VIDEO_PATH,
    "--max-tokens", "500"
], capture_output=True, text=True)

print(result.stdout)
print(result.stderr)
```

```python
#Extract the generated text from the output

#Split by the separator "==========" and take middle part
lines = result.stdout.split("==========")
generated_text = lines[1].strip()

#Get part after "Prompt:" and strip whitespace
generated_text = generated_text.split("Prompt:")[-1].strip()

print(generated_text)
```

```python
#Parse the generated text into a list of salient events
#Assuming each sentence is a seperate event
def parse_salient_events(text):
    events = [e.strip() for e in text.split(". ") if e.strip()]
    return events

events = parse_salient_events(generated_text)
print("(\n" + ",\n".join(events) + "\n)")
#Probably convert this into some .txt file for the next part
```
