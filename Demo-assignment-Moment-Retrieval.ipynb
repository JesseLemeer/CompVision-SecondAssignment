# Demo: Moment Retrieval    

This notebook demonstrates a possible approach for modelling the annotations you have created for the provided videos.

In this case, we focus on designing a moment retrieval module, which takes untrimmed videos as input and localizes when a particular event occurs within the video.

These models aim to retrieve relevant video segments by predicting their start and end timestamps.

As part of this notebook, we demonstrate the use of the Lighthouse repository, which you can also use for your own experiments.


```python
import torch
from lighthouse.models import CGDETRPredictor
```

Using the Lighthouse repository, we will select one of the available models and apply it to the annotated queries.

In this demonstration, we use a widely adopted model for moment retrieval, CG-DETR. More details about the method can be found in the provided documentation.

This example is designed to run on CPU only. The Lighthouse library provides access to several models, which you are encouraged to explore depending on your available computational resources.

```python
model = CGDETRPredictor('clip_cg_detr_qvhighlight.ckpt', device='cpu',
                        feature_name='clip')
```

```python
video = model.encode_video("TVSum/video_5.mp4")
query = 'a truck next to a forest'
```

```python
prediction = model.predict(query, video)
print(f'Closest Moment: {prediction["pred_relevant_windows"][0][0], prediction["pred_relevant_windows"][0][1]}, Corresponding Saliency score: {prediction["pred_relevant_windows"][0][2]}')
```

As we can see, the predicted moment appears to be reasonably accurate. However, what happens when we slightly alter the query? Does the model still capture the moment accurately?

```python
query = 'a crow is playing with a ball close to a door'
prediction = model.predict(query, video)
print(f'Closest Moment: {prediction["pred_relevant_windows"][0][0], prediction["pred_relevant_windows"][0][1]}, Corresponding Saliency score: {prediction["pred_relevant_windows"][0][2]}')
```

This result appears to be less accurate, possibly because the query is too ambiguous. This suggests that the quality and specificity of the query can significantly influence how accurately the model segments a moment in the video. We will now rewrite the query to test whether a clearer description improves retrieval.

```python
query = 'a crow is walking towards a white door'
prediction = model.predict(query, video)
print(f'Closest Moment: {prediction["pred_relevant_windows"][0][0], prediction["pred_relevant_windows"][0][1]}, Corresponding Saliency score: {prediction["pred_relevant_windows"][0][2]}')
```

The prediction is now more accurate in terms of where the event occurs in the video.

Note that the example video used here is only 115 seconds long, while the library supports videos up to 150 seconds in duration. As a result, you may need to make appropriate engineering choices when processing the videos provided to you.

For example, you might consider splitting longer videos into smaller segments or selecting relevant portions before applying the model.
